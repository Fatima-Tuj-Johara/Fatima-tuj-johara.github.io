### Q.1 ) Show the number of Confirmed , Deaths and Recovered cases in each Region.

```python
data.head(2)
```

```python
data.groupby('Region')[['Confirmed','Recovered','Deaths']].sum().head(10)
```

```python
data.groupby('Region')[['Confirmed','Recovered','Deaths']].sum().head(10).plot.bar()
```

### Q2) Remove all the records where Confirmed Cases is Less Than 10.

```python
data.head(2)
```

```python
data[~(data['Confirmed']<10)]
```

### Q. 3) "Which Region had the highest total (cumulative) number of Confirmed cases recorded?"

data.groupby('Region').Confirmed.sum().sort_values(ascending = False).head(20).reset_index()

### Q.4) Which Region had the lowest total number of Deaths recorded? ?

```python
#data.head(2)
data.groupby("Region")['Deaths'].sum().sort_values(ascending=True).head(10).reset_index()
```

### Q.5) How many Confirmed , Deaths & Recovered cases were reported from Bangladesh till 29 April 2020 ?
