### In which year highest number of the TV Shows & Movies were released ? Showing with Bar Graph.

```python
nd.dtypes
```

```python
nd['Date_N']=pd.to_datetime(nd["Release_Date"].str.strip())
```

```python
nd.head(2)
```

```python
nd['Date_N'].dt.year.value_counts()   # It counts the occurrence of all individual Years in Date column.
```

```python
nd['Date_N'].dt.year.value_counts().plot(kind="bar") #bar graph
```
